```python
import requests
import json
import pandas as pd

link = 'https://www.thecocktaildb.com/api/json/v1/1/search.php?s=margarita'
CT = requests.get(link)
```

```python
drink = json.loads(CT.text)
drink = drink['drinks']
drink
```

```python
Margarita = pd.DataFrame(drink)
Margarita.head(6)
```
