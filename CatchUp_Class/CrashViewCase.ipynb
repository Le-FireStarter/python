```python
import requests
import json
import pandas as pd

url = 'https://crashviewer.nhtsa.dot.gov/CrashAPI/crashes/GetCaseList?states=1,51&fromYear=2014&toYear=2015&minNumOfVehicles=1&maxNumOfVehicles=6&format=json'
crash = requests.get(url)
crash.json()
```

```python
crashed = crash.json()['Results'][0]
crashed
```

```python
crash_set = pd.DataFrame(crashed)
crash_set.head(6)
```
